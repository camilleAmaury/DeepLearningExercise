## Imports

```python
import pylab
from matplotlib import pyplot as plt
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
```

## Preprocessing

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
labels = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"]
```

```python
print("""--> x_train :
         - shape : {}
         - data example : \n{}\n""".format(x_train.shape, x_train[100][10]))
print("""--> y_train 
         - shape : {}
         - data example : \n{}\n""".format(y_train.shape, y_train[:10]))
print("""--> x_test :
         - shape : {}
         - data example : \n{}\n""".format(x_test.shape, x_test[100][10]))
print("""-->y_test 
         - shape : {}
         - data example : \n{}\n""".format(y_test.shape, y_test[:10]))
```

Regarding the size of the training and test batches, we can say that we have a \~85-15% proportion for each batch.
Moreover, we should consider a validation batch which will be 6000 rows (\~8.5% of the total datas).

We then, propose to reshape inputs arrays to match tensorflow's pattern. The datas represents a 28*28 pixels image with gray scale (0-255). We will try to normalize all these values between 0 and 1 by dividing by 255, It will contract informations and enable a better understanding for the neural network.

Last but not the least, we will use the tensorflow method which will convert our categories (0-9) to a 10 lengthed vector containing 0 on the not corresponding classes and 1 on which class it corresponds to.

```python
x_train = x_train.reshape(x_train.shape[0], -1) / 255.0
x_test = x_test.reshape(x_test.shape[0], -1) / 255.0
```

```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
print("""--> x_train :
         - shape : {}
         - data example : \n{}\n""".format(x_train.shape, x_train[100][:10]))
print("""--> y_train 
         - shape : {}
         - data example : \n{}\n""".format(y_train.shape, y_train[0]))
print("""--> x_test :
         - shape : {}
         - data example : \n{}\n""".format(x_test.shape, x_test[100][:10]))
print("""-->y_test 
         - shape : {}
         - data example : \n{}\n""".format(y_test.shape, y_test[0]))
```

## Simple Neural Network

In this part we will try to explore the simple to more complex neural network

### Single hidden layer

This model will be the simplier we will be using. We choose the rectified linear unit (reLU) function, since it is quite better for backpropagation and non-linear behavior, applied on the 10 hidden neurons layer.

Then our output layer will be a 10 neurons using the softmax function in order to represents the 10 classes vector with the predicted class.

Lastly, we will be using the categorical crossentropy regarding to our problem, the optimizer could have been "rmsprop" rather than "adam", but here we don't mind. Our comparaison metric will be the accuracy to predict a class.

```python
model = Sequential()
model.add(Dense(10, input_dim=x_train.shape[1], activation='relu'))
model.add(Dense(10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, validation_split=0.1)
```

As we can see, we reached an accuracy of \~0.85, which is quite good for a first try on a validation batch.
Now, we need to test it on the final test batch.

```python
_, test_acc = model.evaluate(x_test, y_test)
print(test_acc)
```

The result seems the same.

We will now explore the repartition of predictions

```python
y_predicted = model.predict(x_test)
cm = np.array(tf.math.confusion_matrix(labels=tf.argmax(y_test, 1), predictions=tf.argmax(y_predicted, 1)))
```

```python
def create_labels(m, length):
    x_labels = []
    for i in range(length):
        x_labels.append("Predicted {}".format(labels[i]))
    y_labels = []
    for i in range(length):
        y_labels.append("Is {}".format(labels[i]))
    return (x_labels, y_labels)

def heatmap_numbers(m, labels, title):
    lenght = len(labels)
    
    # Creating plot base config
    fig, ax = plt.subplots(figsize=(16,10))
    im = ax.imshow(m, cmap="copper")
    
    # Creating labels
    (x_axis_labels, y_axis_labels) = create_labels(cm, lenght)

    # Positionning Labels on axis
    ax.set_xticks(np.arange(lenght))
    ax.set_xticklabels(x_axis_labels)
    ax.set_yticks(np.arange(lenght))
    ax.set_yticklabels(y_axis_labels)
    ax.grid(False)

    # Rotate labels on x axis
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

    # Put values in each case
    for i in range(lenght):
        for j in range(lenght):
            text = ax.text(j, i, m[i, j], ha="center", va="center", color="w")

    ax.set_title(title)
    fig.tight_layout()
    plt.show()
```

```python
heatmap_numbers(cm, labels, "Confusion Matrix of a Single Layer Neural Network on MNSIT Fashion")
```

As we can see, there are some classes which are really harder to predict than the other :
 * Shirt
 * Coat

Else we can consider the network as a pretty good predicter.

### Single increased hidden layer

We will now basically try to increase the number of neurons in the hidden layer.

```python
model2 = Sequential()
model2.add(Dense(50, input_dim=784, activation='relu'))
model2.add(Dense(10, activation='softmax'))
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model2.fit(x_train, y_train, epochs=10, validation_split=0.1)
```

As we can see, we reached an accuracy of \~0.88, which is slighly better on a validation batch.
Now, we need to test it on the final test batch.

```python
_, test_acc = model2.evaluate(x_test, y_test)
print(test_acc)
```

The result seems the same. We can easily conclude that this network is better than the precedent one but not by a huge amount.

We will now explore the repartition of predictions

```python
y_predicted = model2.predict(x_test)
cm = np.array(tf.math.confusion_matrix(labels=tf.argmax(y_test, 1), predictions=tf.argmax(y_predicted, 1)))
heatmap_numbers(cm, labels, "Confusion Matrix of a Single Increased Layer Neural Network on MNSIT Fashion")
```

The problem remains the same that before.

### Multi-hidden Increased Layers

In order to improve our result, we will now give a try to more hidden layers.

```python
model3 = Sequential()
model3.add(Dense(50, input_dim=784, activation='relu'))
model3.add(Dense(50, activation='relu'))
model3.add(Dense(10, activation='softmax'))
model3.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model3.fit(x_train, y_train, epochs=10, validation_split=0.1)
```

As we can see, we reached an accuracy of \~0.90, which is better.
Now, we need to test it on the final test batch.

```python
_, test_acc = model3.evaluate(x_test, y_test)
print(test_acc)
```

The result decreased a lot on the test batch, meaning it could be overfitting or that it won't performs well on new data.

We will now explore the repartition of predictions

```python
y_predicted = model3.predict(x_test)
cm = np.array(tf.math.confusion_matrix(labels=tf.argmax(y_test, 1), predictions=tf.argmax(y_predicted, 1)))
heatmap_numbers(cm, labels, "Confusion Matrix of a Multi Increased Layers Neural Network on MNSIT Fashion")
```

## Convolutional neural network

The CNNetwork performs often well on pattern recognition applied to image. That's why we are going to try it.

In the first place, we need to add a dimension to our inputs : which represents the scale of the color (here we are in grayscale so 1, else would be 3 in RGB).

```python
(x_train, _), (x_test, _) = fashion_mnist.load_data()
x_train = x_train[:,:,:,np.newaxis] / 255.0
x_test = x_test[:,:,:,np.newaxis] / 255.0
```

Then, we create a model following these steps :
* A convolutional layer with a window size of 2*2, creating 64 new smaller images using reLU function. Then we use pooling (here max pooling) to decrease the number of data and keeping (or trying to) the most part of the information.
* Repeat the last steps with smaller images and filters to create more detailled part.
* the last layer as always, is the final vector we want to get the classification.

```python
model4 = Sequential()
model4.add(Conv2D(filters=64, kernel_size=(2,2), padding='same', activation='relu', input_shape=(28,28, 1))) 
model4.add(MaxPooling2D(pool_size=(2,2)))
model4.add(Conv2D(filters=32, kernel_size=(2,2), padding='same', activation='relu', input_shape=(14,14,64))) 
model4.add(MaxPooling2D(pool_size=(2,2)))
model4.add(Flatten())
model4.add(Dense(10, activation='softmax'))
model4.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model4.summary()
```

```python
model4.fit(x_train, y_train, epochs=10, validation_split=0.1)
```

```python
_, test_acc = model4.evaluate(x_test, y_test)
print(test_acc)
```

```python
y_predicted = model4.predict(x_test)
cm = np.array(tf.math.confusion_matrix(labels=tf.argmax(y_test, 1), predictions=tf.argmax(y_predicted, 1)))
heatmap_numbers(cm, labels, "Confusion Matrix of a Convolutional Neural Network on MNSIT Fashion")
```
